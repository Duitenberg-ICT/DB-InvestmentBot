## Price Prediction using Belfort AI (Tutorial)

### Import
At first, we need to import Predictor class from belfort.predictor.

```python
from belfort.predictor import Predictor
```

### Overview and Plot
Then, choose your favourite ticker you want make a prediction for (multiple tickers will be supported later) and period of data (by default 3 years).

```python
ticker = "SOUN"
predictor = Predictor(ticker, period="3y")
```

To plot pulled data, just call plot() or lag_plot() functions like below.

```python
predictor.plot()
```

```python
predictor.lag_plot()
```

You can overview main statistical data such as mean, std, percentiles, and so on by calling overview() function.

```python
predictor.overview()
```

### Prediction
To predict price you can call functions:
- predict(): predicts value for next day, based on moving average, and also returns main statistical data for it. If you want to print it, just put verbose=True (False by default). Has a relatively high accuracy.
- long_predict(): predicts values for the next number of days (by default 365) using Prophet library by Meta. If you want plot graphs with prediction, set verbose=True (False by default). Has lower accuracy then predict() and recommended only for general trend observation.

```python
predictor.predict(verbose=True)
```

```python
test = predictor.long_predict(days=365, verbose=True)
test['trend'].head(90)
```
